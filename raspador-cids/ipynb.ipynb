```python
import fitz
import re
import unidecode

pdf = fitz.open('cids.pdf')

text = ""
for page in pdf:
    text += page.get_text()
    
```

```python
text = text.replace("\xad", " ").replace("\xa0", " ").replace("   ", " ")
text

```

```python
cids = text.split("\n")[5:]
for i, cid in enumerate(cids):
    cids[i] = re.sub(r'\+', ' ', cid)



```

```python
cids_splited = [cid.split(" ", 1) for cid in cids]
for i in range(len(cids_splited)):
    if len(cids_splited[i]) == 1:
        cids_splited[i-1][1] += " " + cids_splited[i][0]
display(cids_splited)
```

```python
cids_splited = [cid for cid in cids_splited if len(cid) == 2]
cids_splited
```

```python
cids_json = [
    
    {
        "cid": cid[0],
        "descricao": cid[1],
        "unidecode": unidecode.unidecode(cid[1]).lower()
    }
    
    for cid in cids_splited
]

```

```python
import json


with open("cids.json", "w") as f:
    json.dump(cids_json, f)
```
